# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
%matplotlib inline
```

# Loading Data

```python
data = pd.read_csv('data.csv')
data
```

# Splitting Data to 12 Subjects

```python
subjects = data["Subject"].apply(lambda x: int(x.split(".")[0])).unique() # example: 1.mat -> 1
subjects_ordered = np.sort(subjects)
subjects_ordered
```

```python
dataframes = []

for s in subjects_ordered:
    df = data[data["Subject"] == str(s) + ".mat"] # example: 1 + .mat -> 1.mat
    dataframes.append(df)
```

Now we have an array of 12 different dataframes for 12 different subjects ordered from 1->12. Example:

```python
dataframes[0] #1.mat
```

```python
plt.figure(figsize=(15, 6))
plt.ylim([-100, 100])
plt.plot(dataframes[0]["Unnamed: 0"][1500:100000], dataframes[0]["HGRF"][1500:100000])
plt.title("HGRF")
```

# EDA

## Univariate Analysis

### Descriptive Statistics

```python
features = ["HGRF", "LGRF", "VGRF"]
for f in features:
    print(data[f].describe())
```

### Box-plots & Outliers

```python
for f in features:
    print(f)
    plt.figure()
    data.boxplot(column = [f])
    plt.show()
```

It's obvious that HGRF & LGRF have some extreme outliers.

### Skewness & Kurtosis

```python
from scipy.stats import skew

for f in features:
    print(f)
    print(skew(data[f], axis=0))
```

```python
from scipy.stats import kurtosis

for f in features:
    print(f)
    print(kurtosis(data[f], axis=0))
```

HGRF & VGRF can be considered normally distributed, while LGRF can't since it's not within -2 < kurtosis < +2.

### Line Plots & Variation

```python
plt.rcParams['figure.figsize']

for f in features:
    plt.figure(figsize=(15, 6))
    plt.plot(data["Unnamed: 0"], data[f])
    plt.title(f)
    plt.show()
```

## Multivariate Analysis

### Correlation

```python
data[["HGRF", "LGRF", "VGRF"]].corr()
```

```python
cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(data[["HGRF", "LGRF", "VGRF"]].corr(), cmap = cmap, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Looks like there is no strong correlations or multi-collinearity that might affect our future model.

### 3D Scatter Plot

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data["HGRF"], data["VGRF"], data["LGRF"],
           edgecolor='k')

ax.set_xlabel('HGRF', labelpad=20)
ax.set_ylabel('VGRF', labelpad=20)
ax.set_zlabel('LGRF', labelpad=20)

plt.show()
```
